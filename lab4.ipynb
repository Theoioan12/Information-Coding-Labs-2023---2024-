### 1 Eratosthenes sieve
Define a function ```Eratosthenes``` implementing Eratosthenes sieve.

INPUT: a positive integer n

OUTPUT: a list of primes up to n, generated by sieving n


````python
def eratosthenes(n):
    prime = [True for _ in range(n + 1)]
    p = 2
    while p**2 <= n:
        if prime[p] is True:
            for i in range(p**2, n + 1, p):
                prime[i] = False
        p += 1

    primes = [p for p in range(2, n + 1) if prime[p]]

    return primes

n = 30 
result = eratosthenes(n)
print(result)
````

### 2 Trial division
Define a function ```Trial``` implementing the trial division algorithm.

INPUT: an integer

OUTPUT: a boolean. True if p is prime, false if not. If p is not prime, print its smallest divisor.


````python
def trial(p):
    if p < 2:
        return False

    for i in range(2, int(p**0.5) + 1):
        if p % i == 0:
            print(f"{p} is not prime. Smallest divisor: {i}")
            return False

    print(f"{p} is prime.")
    return True


number_to_check = 17
print(trial(number_to_check))
number_to_check = 18
print(trial(number_to_check))
number_to_check = 89
print(trial(number_to_check))
number_to_check = 77
print(trial(number_to_check))
````

### 3 Miller-Rabin
Define a function ```MillerRabin``` implementing the Miller-Rabin primality test.

INPUT: an integer n, an error bound epsilon

OUTPUT: a boolean (false if n fails one of the Miller tests and it's not prime,
            true if n passes all the tests and could be prime).
            
If  p is not prime, print the last base used as certificate.
	
*Remark: although the Miller test has to be repeated several times, the computation of $s$ and $t$
such that $n-1=(2^s)t$ only has to be made once.*

````python

import math
import random 
 
def power(x, y, p):
     
    res = 1; 
    x = x % p; 
    while (y > 0):
        if (y & 1):
            res = (res * x) % p;
 
        y = y>>1; # y = y/2
        x = (x * x) % p;
     
    return res;

def miillerTest(d, n):
     
    a = 2 + random.randint(1, n - 4);

    x = power(a, d, n);
 
    if (x == 1 or x == n - 1):
        return True;

    while (d != n - 1):
        x = (x * x) % n;
        d *= 2;
 
        if (x == 1):
            return False;
        if (x == n - 1):
            return True;
 
    return False;
 
def isPrime( n, k):
    if (n <= 1 or n == 4):
        return False;
    if (n <= 3):
        return True;

    d = n - 1;
    while (d % 2 == 0):
        d //= 2;
 
    for i in range(k):
        if (miillerTest(d, n) == False):
            return False;
 
    return True;

def generate_miller(n, epsilon):
    # Driver Code
    # Number of iterations
    k = -math.log(epsilon, 4); 
    k = math.ceil(k)
    numbers = []
    for n in range(1,n):
        if (isPrime(n, k)):
            numbers.append(n)
    return numbers
    

print(generate_miller(100, 0.0001))


errors = 0
for i in range(100):
    if(isPrime(3215031751, 1)):
        errors += 1
        
print(errors)
````
